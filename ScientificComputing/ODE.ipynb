I used the Julia package `DifferentialEquations.jl` to solve an ordinary differential equation and plotted its solution.

```julia
using DifferentialEquations
using Plots
```

```julia
# Formulate the ODE
f(u,p,t) = (t^2)*(sin(t))
u0 = 2
tspan = (1.0,100.0);
prob = ODEProblem(f,u0,tspan);
```

```julia
# Solve the ODE
sol = solve(prob,reltol=1e-8,abstol=1e-8);
```

```julia
# Plot the solution
plot(sol,linewidth=3,title="Algorithmic Solution to the ODE",
     xaxis="Value of t",yaxis="Value of u(t)",legend=false)
```

```julia
# Plot my solution obtained (manual computation - integration by parts)
mySol = t->2*t*sin(t)-((t^2)-2)cos(t); # my calculation leads to this answer
plot(mySol,1.0,100.0,linewidth=3,title="Manual Solution via Integration by Parts",
     xaxis="Value of t",yaxis="Value of u(t)",legend=false)
```
