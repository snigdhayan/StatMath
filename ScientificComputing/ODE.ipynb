I used the Julia package `DifferentialEquations.jl` to solve an ordinary differential equation and plotted its solution.

```julia
using DifferentialEquations

f(u,p,t) = (t^2)*(sin(t))
u0 = 0
tspan = (1.0,100.0);
```

```julia
prob = ODEProblem(f,u0,tspan);
```

```julia
sol = solve(prob,Tsit5(),reltol=1e-8,abstol=1e-8);
```

```julia
using Plots
plot(sol,linewidth=3,title="Solution to the linear ODE",
     xaxis="Variable t",yaxis="Value of u(t)",legend=false)
```
