I used the Julia package `DifferentialEquations.jl` to solve an ordinary differential equation and plotted its solution. I also plotted the solution obtained by manual computation (integration by parts) for comparison.

```julia
using DifferentialEquations
using Plots
```

```julia
# Formulate the ODE
f(u,p,t) = (t^2)*(sin(t))
u0 = 2
tspan = (0.0,100.0);
prob = ODEProblem(f,u0,tspan);
```

```julia
# Solve the ODE
sol = solve(prob,reltol=1e-8,abstol=1e-10);
```

```julia
# Plot the solution
plot(sol,
     xticks=0.0:20.0:100.0,
     linewidth=3,
     title="Algorithmic Solution",
     xaxis="Value of t",
     yaxis="Value of u(t)",
     legend=false)
```

```julia
# Plot my solution (manual computation - integration by parts)
mySol = t->2*t*sin(t)-((t^2)-2)cos(t); # manual computation leads to this answer
plot(mySol,0.0,100.0,
     xticks=0.0:20.0:100.0,
     linewidth=3,
     title="Theoretical Solution",
     xaxis="Value of t",
     yaxis="Value of u(t)",
     legend=false)
```

Observe that the two solutions are quite similar. The solution obtained by manual computation (integration by parts) is the exact solution. The algorithmic solution approximates it.
