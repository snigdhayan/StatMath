We are often interested in forecasting values of a time series dataset. The `ARIMA` model is quite well suited for this purpose. I collected `Google Trends` web search data for the programming languages `Python`, `R` and `Golang` for the past 5 years and used the `forecast` package to forecast the trend with confidence intervals.

```r
# Load 'forecast' package
library(forecast)
```

```r
# Read and visualize the data
df <- read.csv("Python_R_Golang.csv")
head(df, 10)
```

```r
# Global parameter
freq <- 52 # approximate number of weeks in a year
```

```r
# ARIMA model for Python trend
timeseries <- ts(df$Python, start = c(2016, 24), end = c(2021, 24), frequency = freq)

fitModel <- arima(timeseries, order = c(1, 1, 1))
Forecast <- forecast(fitModel, h = 28)
autoplot(Forecast)
```

```r
# look at forecast
Forecast
```

```r
# ARIMA model for R trend
timeseries <- ts(df$R, start = c(2016, 24), end = c(2021, 24), frequency = freq)

fitModel <- arima(timeseries, order = c(1, 1, 1))
Forecast <- forecast(fitModel, h = 28)
autoplot(Forecast)
```

```r
# look at forecast
Forecast
```

```r
# ARIMA model for Golang trend
timeseries <- ts(df$Golang, start = c(2016, 24), end = c(2021, 24), frequency = freq)

fitModel <- arima(timeseries, order = c(1, 1, 1))
Forecast <- forecast(fitModel, h = 28)
autoplot(Forecast)
```

```r
# look at forecast
Forecast
```
