We are often interested in forecasting values of a time series dataset. The `ARIMA` model is quite well suited for this purpose. I collected `Google Trends` web search data for the programming languages `Python`, `R` and `Golang` for the past 5 years and used the `forecast` package to forecast the trend with confidence intervals.

```r
# Load 'forecast' package
library(forecast)
```

```r
# Read and visualize the data
df <- read.csv("Python_R_Golang.csv")
head(df, 10)
```

```r
# Global parameter
freq <- 52 # approximate number of weeks in a year
```

```r
# ARIMA model for Python trend
timeseries1 <- ts(df$Python, start = c(2016, 24), end = c(2021, 24), frequency = freq)

fitModel1 <- auto.arima(timeseries1)
Forecast1 <- forecast(fitModel1, h = 28)
autoplot(Forecast1)
```

```r
# look at forecast
Forecast1
```

```r
# ARIMA model for R trend
timeseries2 <- ts(df$R, start = c(2016, 24), end = c(2021, 24), frequency = freq)

fitModel2 <- auto.arima(timeseries2)
Forecast2 <- forecast(fitModel2, h = 28)
autoplot(Forecast2)
```

```r
# look at forecast
Forecast2
```

```r
# ARIMA model for Golang trend
timeseries3 <- ts(df$Golang, start = c(2016, 24), end = c(2021, 24), frequency = freq)

fitModel3 <- auto.arima(timeseries3)
Forecast3 <- forecast(fitModel3, h = 28)
autoplot(Forecast3)
```

```r
# look at forecast
Forecast3
```
